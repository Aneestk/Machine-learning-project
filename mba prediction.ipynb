## 4. Model Development

#### Splitting dependant and independant features.

```python
x = df.drop('Decided to Pursue MBA?', axis=1)  # Replace 'Decided to Pursue MBA?' with your target variable
y = df['Decided to Pursue MBA?']
```

### 4.1 Splitting the Data
#### The dataset was split into 80% for training and 20% for testing using train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### 4.2 Model selection, Training and Evaluation
* K-Nearest Neighbors (KNN)
* Naive Bayes Classification
* Decision Tree Algorithm
* Random Forest Classifier
* Logistic Regression

### 1)KNN

```python
#import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
classifier_knn=KNeighborsClassifier(n_neighbors=5)
classifier_knn.fit(x_train,y_train)  #training
y_pred_knn=classifier_knn.predict(x_test)
```

```python
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
result=confusion_matrix(y_test,y_pred_knn)
print('Confusion Matrix:')
print(result)
```

```python
result2=accuracy_score(y_test,y_pred_knn)
print('Accuracy:',result2)
```

### 2)NIEVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
model_nb=GaussianNB()
model_nb.fit(x_train,y_train)   #training
y_pred_nb=model_nb.predict(x_test)
```

```python
result3=accuracy_score(y_test,y_pred_nb)
print('Accuracy:',result3)
```

### 3) DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()  #by defualt gini
model_dt=model_dt.fit(x_train,y_train) #training
y_pred_dt=model_dt.predict(x_test)
```

```python
result4=accuracy_score(y_test,y_pred_dt)
print('Accuracy:',result4)
```

```python
# print the score on training and test set

print("Training set score:{:.4f}".format(model_dt.score(x_train,y_train)))
print("Test set score :{:.4f}".format(model_dt.score(x_test,y_test)))
```

```python
model_dt1=DecisionTreeClassifier(criterion="entropy",max_depth=3)
model_dt1=model_dt1.fit(x_train,y_train) #predict the response for the test
y_pred_dt1=model_dt1.predict(x_test) #model accuracy , how often is the classifier
result5=accuracy_score(y_test,y_pred_dt1)
print('Accuracy:',result5)
```

### 4)RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier  #import decision tree classifier
model_rf=RandomForestClassifier(n_estimators=57,random_state=42)  
#train decision tree classifier
model_rf=model_rf.fit(x_train,y_train)
y_pred_rf=model_rf.predict(x_test)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_rf))
```

### 5)LOGISTIC REGRESSION

```python
#fitting logistic regression to the training set
from sklearn.linear_model import LogisticRegression #(building model)
model_lr=LogisticRegression(random_state=42)
model_lr.fit(x_train,y_train)
y_pred_lr=model_lr.predict(x_test)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_lr))
```

### 4.3 Final Model
### the best model (Naive Bayes classifier,Random Forest,Logistic Regression) was retrained on the entire dataset.

```python
print('Accuracy of KNN Classifier:',result2)
print('Accuracy of Naive Bayes classifier:',result3)
print('Accuracy of Decision Tree Algorithm:',result4)
print('Accuracy of Random Forest:',accuracy_score(y_test,y_pred_rf))
print('Accuracy of Logistic Regression:',accuracy_score(y_test,y_pred_lr))
```

# Conclusion

#### Among the five classifiers tested, Naïve Bayes and Logistic Regression achieved the highest accuracy (0.5945), making them the most effective models for this dataset. The Decision Tree classifier performed the worst (0.5015), indicating potential overfitting or inefficiency in learning from the data. Random Forest (0.559) showed an improvement over the Decision Tree, demonstrating the benefits of ensemble learning. KNN (0.5315) performed moderately but might require hyperparameter tuning and feature scaling for better results.

#### For further improvement, techniques such as hyperparameter tuning, feature engineering, and cross-validation should be explored. Depending on the problem's requirements (e.g., interpretability vs. complexity), Logistic Regression or Naïve Bayes may be the best choices
